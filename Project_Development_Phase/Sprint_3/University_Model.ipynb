# SPLITTING DATA INTO TRAIN AND SPLIT

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state = 10)
```

```python
y_train = (y_train > 0.5)
y_train
```

# MODEL BUILDING


# TRAINING AND TESTING THE MODEL

```python
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFE
cls = LogisticRegression(random_state = 0, max_iter=1000,solver = 'lbfgs')
```

```python
#y1_tarin = pd.DataFrame(y_train.value)
print(y_train.shape)
y1_train = np.reshape(y_train,320)
print(y1_train.shape)
lr = cls.fit(x_train, y1_train)
```

```python
y_pred = lr.predict(x_test)
y_pred
```

```python
print(y_test.shape)
y1_test = np.reshape(y_test,80)
print(y1_test.shape)
y1_test
```

# MODEL EVALUATION

```python
from sklearn.metrics import accuracy_score, recall_score, roc_auc_score,confusion_matrix
print("Accuracy score : "+str(accuracy_score(y1_test,y_pred)*100))
print("Recll score : "+str(recall_score(y1_test,y_pred)*100))
print("Roc score : "+str(roc_auc_score(y1_test,y_pred)*100))
print(confusion_matrix(y1_test,y_pred))            
```
